OÂ **Selenium tem vÃ¡rios mÃ©todos pÃºblicos**, como oÂ `find_element`Â que usamos anteriormente e oÂ `find_elements`Â (no plural), utilizados para localizar o primeiro elemento correspondente ao resultado do filtro ou todos os elementos que se encaixarem na busca, respectivamente.

TambÃ©m podemos utilizar oÂ `By`Â para especificar um elementoÂ `CSS`Â ouÂ `XPATH`Â que serÃ¡ buscado, para isso Ã© necessÃ¡rio importÃ¡-lo via:

Nesse caso, devemos passar dois parÃ¢metros: o primeiro parÃ¢metro define o que vocÃª irÃ¡ buscar e o segundo o filtro da nossa pesquisa. Abaixo temos duas opÃ§Ãµes no que diz respeito ao que estamos buscando, uma delas utilizando oÂ `By`Â e a outra no formato previamente utilizadoÂ _(quando buscamos pelo campo de nomeÂ `q`Â no exemplo acima)_.

|Com o By|Sem o By|
|---|---|
|By.ID|â€œidâ€|
|By.NAME|â€œnameâ€|
|By.XPATH|â€œxpathâ€|
|By.LINK_TEXT|â€œlink textâ€|
|By.PARTIAL_LINK_TEXT|â€œpartial link textâ€|
|By.TAG_NAME|â€œtag nameâ€|
|By.CLASS_NAME|â€œclass nameâ€|
|By.CSS_SELECTOR|â€œcss selectorâ€|

Assim como quando utilizamos a libÂ `requests`, inspecionar a pÃ¡gina que queremos raspar Ã© imprescindÃ­vel para entendermos a estrutura de como a pÃ¡gina foi montada e quais elementos devemos selecionar para buscar as informaÃ§Ãµes que queremos.

> scrape_selenium.py

Copiar

```python
from selenium import webdriver

# Importa o By
from selenium.webdriver.common.by import By

firefox = webdriver.Firefox()

firefox.get("https://books.toscrape.com/")


# Define a funÃ§Ã£o que farÃ¡ o scrape da URL recebida
def scrape(url):
    firefox.get(url)

    # Itera entre os elementos com essa classe
    for book in firefox.find_elements(By.CLASS_NAME, "product_pod"):
        # Cria dict vazio para guardar os elementos capturados
        new_item = {}

        # Cria uma chave 'title' no dict que vai receber o resultado da busca
        # O tÃ­tulo estÃ¡ em uma tag anchor que estÃ¡ dentro de uma tag 'H3'
        new_item["title"] = (
            book.find_element(By.TAG_NAME, "h3")
            .find_element(By.TAG_NAME, "a")
            .get_attribute("innerHTML")
        )

        # O preÃ§o do book estÃ¡ em um elemento da classe 'price_color'
        new_item["price"] = book.find_element(
            By.CLASS_NAME, "price_color"
        ).get_attribute("innerHTML")

        # O link estÃ¡ dentro de um atributo 'href'
        new_item["link"] = (
            book.find_element(By.CLASS_NAME, "image_container")
            .find_element(By.TAG_NAME, "a")
            .get_attribute("href")
        )

        print(new_item)


scrape("https://books.toscrape.com/")
```

Utilizamos bastante no cÃ³digo acima o mÃ©todoÂ `get_attribute`. A razÃ£o disso Ã© que com o Selenium, o elemento retornado depois da busca pelo atributoÂ `CSS`Â serÃ¡ do tipo webdriver e nÃ£o texto. Ã‰ justamente para fazer essa conversÃ£o que utilizamos esse mÃ©todo especificando o atributo â€˜innerHTMLâ€™ ou â€˜hrefâ€™, por exemplo.

TambÃ©m utilizamos o mÃ©todoÂ `find_element`Â encadeado para procurar um elemento dentro de outro elemento, como fizemos para pegar o link, por exemplo, que era um elemento Ã¢ncora dentro de umaÂ `div`.
